```python
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
base="https://ko.wikisource.org"
req = requests.get(url)
req.encoding = 'utf-8'

bs = BeautifulSoup(req.text, "html.parser")
work_list = bs.select('div.mw-parser-output > ul > li')

works=[]
for work in work_list:
    works.append([work.a.string, content_out(base+work.a['href'])])
    dwork_list = work.select('li>a')
    for dwork in dwork_list[1:]:
        works.append([dwork.string, content_out(base+dwork['href'])])
works
```

```python
def content_out(url):
    res = requests.get(url)
    res.encoding='utf-8'
    
    bs = BeautifulSoup(res.text, 'html.parser')
    content = bs.select('div.mw-parser-output p')
    str=''
    for cont in content:
        if re.findall('^이 저작물은',cont.text):
            continue
        str+=cont.text
    return str
```

```python
url = 'https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9D%B4%EC%83%81'
res = requests.get(url)
res.encoding='utf-8'
bs=BeautifulSoup(res.text,'html.parser')
link_list = bs.select('div.mw-parser-output > ul:nth-child(9) a')

base='https://ko.wikisource.org'

result=[]

for link in link_list:
    title = link['title']
    content = get_content(base+link['href'])
    year = link.next_sibling.strip()[1:-1].split(',')[0]
    result.append([title,year,content])
print(result)
```

```python
def get_content(url):
    res = requests.get(url)
    res.encoding='utf-8'
    bs=BeautifulSoup(res.text, 'html.parser')
    p_list = bs.select('div.mw-parser-output p')
    content=''
    for p in p_list:
        if re.findall('^이 저작물은',p.text):
            continue
        content+=p.text
    return content
```
