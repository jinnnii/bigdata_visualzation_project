```python
#2차원 리스트를 만들면, 행단위로 가져오는 건 되지만, 열 단위로 슬라이싱이 되지 않는다!
# Numpy 는 열단위로 슬라이싱이 가능
list1=[[1,11], [2,22],[3,33]]

arr4 = np.array(list1)
print(arr4[1,1])  #22
print(arr4[1][1]) #22

print(arr4[:,1]) #[11 22 33] 모든 행의 1열
print(arr4[1,:]) #[ 2 22]    1행의 모든 열
```

```python
list=[5,3,4,2,5,1,7,6,9]
arr=np.array(list).reshape(3,3)

print(arr)
print(np.sort(arr)) #행 기준 정렬
print(np.sort(arr,axis=1)) #행 기준 정렬
print(np.sort(arr,axis=0)) #열 기준 정렬

print(np.sort(arr,axis=0)[::-1]) #열 기준 역순 정렬
print(np.sort(arr,axis=1)[::,::-1]) #행 기준 역순 정렬
```

```python
print(np.linspace(0,100,5)) #0부터 100사이에서 동일 간격의 4개의 값으로 생성
print(np.linspace(0,100,5,retstep=True)) # retstep = 간격이 얼마인지 추가 정보
print(np.linspace(0,100,5,endpoint=True)) # 마지막 값을 포함(또는 미포함)
print(np.linspace(0,100,5,endpoint=False))

print(np.logspace(1,2,4)) #log10_10 log10_100 까지 4등간격으로 값 생성
```
