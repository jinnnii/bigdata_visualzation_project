```python
import requests
import pandas as pd
r = requests.get('https://movie.naver.com/movie/point/af/list.naver')
r.encoding='utf-8'
bs = BeautifulSoup(r.text, 'html.parser') 
title_list = bs.find_all(class_='title')
for i in title_list:
    title = i.find('a').text
    score = i.find('em').text
    content = i.br.next_sibling.text.strip()
    author=i.next_sibling.next_sibling.a.text
    date = i.next_sibling.next_sibling.br.next_sibling.text
    print({'title':title,'score':score,'content':content, 'author':author, 'date':date})
```

```python
url= 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
req = requests.get(url)
req.encoding='utf-8'

bs = BeautifulSoup(req.text, 'xml')

location_list = bs.find_all('location')
for location in location_list:
    data_list = location.find_all('data')
    print(location.find('city').string)
    for data in data_list:
        mode = data.find('mode').string
        date, time = data.find('tmEf').string.split()
        wf =  data.find('wf').string
        tmn = data.find('tmn').string
        tmx = data.find('tmx').string
        rnst = data.find('rnSt').string
        print([mode,date,time,wf,tmn,tmx,rnst])
```

```python
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
base="https://ko.wikisource.org"
req = requests.get(url)
req.encoding = 'utf-8'

bs = BeautifulSoup(req.text, "html.parser")
work_list = bs.select('div.mw-parser-output > ul > li')

works=[]
for work in work_list:
    works.append(work.a.string)
    print(base+work.a['href'])
    dwork_list = work.select('li>a')
    for dwork in dwork_list[1:]:
        print(base+dwork['href'])
        works.append(dwork.string)
works
```

```python
url = "https://www.pythonscraping.com/pages/warandpeace.html"
req = requests.get(url)
req.encoding='utf-8'

bs = BeautifulSoup(req.text, 'html.parser')
green_tags = bs.select('.green')
green_words = [word.string.replace('\n',' ') for word in green_tags]
green_words
```

```python
url = "https://www.pythonscraping.com/pages/page3.html"
req = requests.get(url)
req.encoding='utf-8'

bs = BeautifulSoup(req.text, 'html.parser')

gift_list = bs.find('table',id='giftList').find_all('tr',class_='gift')
for gift in gift_list:
    title, desc, cost, img = gift.find_all('td')
    print([title.string.strip(), desc.text.strip(),cost.string.strip(),img.img['src']])
```
