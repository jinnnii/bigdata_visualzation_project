```python
import pandas as pd

red_df = pd.read_csv('data/winequality-red.csv', sep=';', header=0)
white_df= pd.read_csv('data/winequality-white.csv',sep=';',header=0)

red_df.to_csv('data/red-wine.csv',index=False)
white_df.to_csv('data/whitewine.csv',index=False)
```

```python
red_df
# quality : 등급
# other: 레드와인 퀄리티를 결정하는 요소
```

```python
# 레드와인과 화이트와인 합치기
# type 컬럼 생성 하여 red, white로 구분 입력
red_df.insert(0, column='type', value='red')
```

```python
white_df.insert(0, column='type', value='white')
```

```python
wine= pd.concat([red_df, white_df])
wine
```

```python
wine.to_csv('data/wine.csv',index=False)
```

```python
wine.columns=wine.columns.str.replace(' ','_')
wine.head()
```

```python
# import scipy.stats
from scipy import stats
from statsmodels.formula.api import ols, glm

red_wine_q = wine.loc[wine['type']=='red','quality']
white_wine_q=wine.loc[wine['type']=='white','quality']

# scipy 패키지의 stats.ttest_ind : t-검정 수행, 두 그룹간 차이 확인
stats.ttest_ind(red_wine_q,white_wine_q, equal_var =False)
```

```python
#선형 회귀 분석식의 종속 변수(y)와 독립변수 (x1~x10)를 구성

#퀄리티(y)는 다른 독립변수들(x1~x10)에 영향을 받는다.
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
      residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + \
      density + pH + sulphates + alcohol'
# 선형 회귀 모델 중에서 OLSOrdinary Least Squares 모델 사용
regresstion_result = ols(Rformula, data=wine).fit()

# 선형 회귀 분석과 관련된 통계값 출력
regresstion_result.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#커널 밀도 추정(kde)를 적용한 히스토그램 그리기
#히스토그램 차트의 배경 스타일 설정
sns.set_style('dark')
#레드와인에 대한 distplot 객체 생성
sns.histplot(red_wine_q,kde=True,color='red', label='red wine')
#화이트와인에 대한 distplot 객체 생성
sns.histplot(white_wine_q,kde=True, label='white wine')
plt.title('Quality of Wine Type')

plt.legend()
plt.show()
```

```python
#부분 회귀 계산을 위해 statsmodels.api 로드
import statsmodels.api as sm

#fixed_acidity가 종속변수 quality에 미치는 영향력 시각화

#부분  회귀에 사용된 독립변수와 종속변수를 제외한 나머지 변수 이름을 리스트 others 로 추출
others= list(set(wine.columns).difference(set(['quality','fixed_acidity'])))
p,resids = sm.graphics.plot_partregress('quality','fixed_acidity', others, data=wine,
                                        ret_coords=True)
plt.show()
```
