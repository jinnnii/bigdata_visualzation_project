```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', loc='right')
plt.ylabel('Y-Axis', loc='top')
# plt.grid(True)
# plt.grid(True,axis='x')
title = plt.title('Graph Title', loc='right', pad=20)
title_str= title.get_text()

plt.axhline(5, 0.0,1.0, color='gray', linestyle='-', linewidth=3)
plt.hlines(8,1.5, 3.0, color='red', linestyle='solid', linewidth=3)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10],label='Price($)')
plt.plot([1, 2, 3, 4], [3,5,9,7],label='Demand(#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend(loc=(0.0,0.0))
plt.legend(loc=(1.0,1.0))
plt.legend(loc='lower right')
plt.legend(loc='best', ncol=2)
plt.legend(fontsize=14)

plt.legend(frameon=True, shadow=True)
plt.show()
```

```python

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])      # X : [xmin, xmax]
plt.ylim([0, 20])     # Y : [ymin, ymax]
plt.axis([0,5,0,20])
plt.show()
```

```python
x=[1,2,3,4]
y1=[2,3,5,10]
y2=[1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill_between(x[1:3], y1[1:3],y2[1:3],color='gray',alpha=0.5)
plt.show()
```
