```python
from folium import plugins
import os

#클러스터링

N=100
data=np.array(
    [np.random.uniform(low=35, high=38, size=N), #Random latitudes
     np.random.uniform(low=125,high=129,size=N), #Random longitudes
    ]
).T
popups=[str(i) for i in range(N)] #Popups texts are simple numbers
m=folium.Map([data[0][0], data[0][1]], zoom_start=8)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m
```

```python
import pandas as pd
cb = pd.read_csv('data/coffeebean.csv', encoding='CP949', index_col=0, header=0)
cb.head()
```

```python
addr=[]
for a in cb.address:
    addr.append(a.split())
addr
```

```python
#행정 구역 표준 이름이 아닌 거 수정하기

addr_list=[]
for i,v in enumerate(addr):
    v=v[0]
    
    if v=='서울':addr[i][0]='서울특별시'
    elif v=='서울시' : addr[i][0]='서울특별시'
    elif v=='부산시' : addr[i][0]='부산광역시'
    elif v=='인천' : addr[i][0]='인천광역시'
    elif v=='광주' : addr[i][0]='광주광역시'
    elif v=='대전시' : addr[i][0]='대전광역시'
    elif v=='울산시' : addr[i][0]='울산광역시'
    elif v=='세종시' : addr[i][0]='세종특별자치시'
    elif v=='경기' : addr[i][0]='경기도'
    elif v=='충북' : addr[i][0]='충청북도'
    elif v=='충남' : addr[i][0]='충청남도'
    elif v=='전북' : addr[i][0]='전라북도'
    elif v=='전남' : addr[i][0]='전라남도'
    elif v=='경북' : addr[i][0]='경상북도'
    elif v=='경남' : addr[i][0]='경상남도'
    elif v=='제주' : addr[i][0]='제주특별자치도'
    elif v=='제주도' : addr[i][0]='제주특별자치도'
    elif v=='제주시' : addr[i][0]='제주특별자치도'
    
    addr_list.append(' '.join(addr[i]))
addr_list
```

```python
addr_df = pd.DataFrame(addr_list, columns=['address2'])
addr_df
```

```python
# cb2 = pd.concat([cb,addr_df],axis=1)
cb['address2']=addr_list
cb.head()
```

```python
cb.to_csv('data/CoffeeBean_2.csv', encoding='CP949', index=False)
```
