```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc, font_manager

sns.set(style='darkgrid')

%matplotlib inline

# 한글 처리를 위한 폰트 설정
font_name= font_manager.FontProperties(fname='c:/windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)

# 그림 크기 설정
plt.rcParams['figure.figsize']=(12,9)

# 마이너스 표기
plt.rcParams['axes.unicode_minus']=False

# 타이타닉 데이터 로드
titanic= sns.load_dataset('titanic')

# Tips 데이터 로드
tips=sns.load_dataset('tips')
```

```python
sns.catplot(x='day', y='tip', data=tips)
#요일별 요금 정산 금액
# 토요일에 가장 높은 
```

```python
# 펼쳐서 겹침을 없애 줌, 정확한 개수를 알 수 있음
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
# 색상 설정
sns.catplot(x='day', y='total_bill',data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill',data=tips, kind='violin')
```

```python
#기본 violinplot 그리기
sns.violinplot(x=tips['total_bill'])
```

```python
#x,y 축을 지정해줌으로써 바이올린을 분할하여 비교 분포를 볼 수 있다.
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#violinplot의 하이라이트로, split 옵션으로 바이올린을 합쳐서 볼 수 있다.

sns.violinplot(x='day', y='total_bill', hue='smoker',
               data=tips, palette='muted', split=True)
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='violin',
            hue='smoker', inner='stick', split=True)

#split : 중앙선으로 카테고리 성별로 나누어짐
#inner=stick : 중앙에 선 그림(밀도)
#실제로 영수금액이 일어난 위치 (몰려있을 수록 많이 일어남)
```

```python
# 박스 중앙에 
# . : 정말 가끔씩 나오는 부분
# ㅡ : 최솟값, 최댓값
# 네모안의 ㅡ  : 값이 가장 많이 분포하고 있는 부분

sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
#박스보다 분포도를 좀더 상세히
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
# 실선 : 오차범위
```

```python
sns.catplot(x='day', data=tips, kind='count')
```

```python
sns.catplot(x='sex', y='survived', data=titanic, 
            hue='class', kind='bar')
```

```python
sns.pairplot(tips)

#자기자신은 히스토그램
#그외에는 두 데이터간 관계를 보여주는 분포도
```

```python
sns.pairplot(tips, hue='size')
```

```python
#lmplot
sns.lmplot(x='total_bill', y='tip', height=8, data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', 
           hue='smoker',height=8, data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker',
           col='day', col_wrap=2, height=6, data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', height=6, data=tips)
```

```python
#선형 관계를 표현하는 regresstion 라인그리기 가능

sns.jointplot(x='total_bill', y='tip', 
              height=6, data=tips, kind='reg')
```
